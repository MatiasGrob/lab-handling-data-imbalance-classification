```python
X_train_categorical = X_train.select_dtypes(object)
X_train_numerical = X_train.select_dtypes(np.number)
X_test_categorical = X_test.select_dtypes(object)
X_test_numerical = X_test.select_dtypes(np.number)
```

```python
from sklearn.preprocessing import MinMaxScaler 

transformer = MinMaxScaler().fit(X_train_numerical)
X_train_norm = transformer.transform(X_train_numerical)
print(X_train_norm.shape)
X_train_numerical_norm = pd.DataFrame(X_train_norm, columns=X_train_numerical.columns)
X_train_numerical_norm.head()
```

```python
transformer = MinMaxScaler().fit(X_test_numerical)
X_test_norm = transformer.transform(X_test_numerical)
print(X_test_norm.shape)
X_test_numerical_norm = pd.DataFrame(X_test_norm, columns=X_test_numerical.columns)
X_test_numerical_norm.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='first').fit(X_train_categorical)

cols = encoder.get_feature_names(input_features=X_train_categorical.columns)
X_train_categorical_encode = pd.DataFrame(encoder.transform(X_train_categorical).toarray(),columns=cols)
X_train_categorical_encode.head()
```

```python
encoder = OneHotEncoder(drop='first').fit(X_test_categorical)

cols = encoder.get_feature_names(input_features=X_test_categorical.columns)
X_test_categorical_encode = pd.DataFrame(encoder.transform(X_test_categorical).toarray(),columns=cols)
X_test_categorical_encode.head()
```

```python
X_train_concat=pd.concat([X_train_numerical_norm, X_train_categorical_encode], axis=1)
X_test_concat=pd.concat([X_test_numerical_norm, X_test_categorical_encode], axis=1)
```

```python
X_train_concat.shape
```

```python
X_test_concat.shape
```

```python
from sklearn.linear_model import LogisticRegression

classification = LogisticRegression(random_state=0, solver='lbfgs', max_iter=1000, 
                 multi_class='multinomial').fit(X_train_concat, y_train)
```

```python
classification
```

```python
from sklearn.linear_model import LinearRegression as LinReg
from sklearn import linear_model

linreg = linear_model.LinearRegression()
linreg
```

```python
linreg.fit(X_train_concat, y_train)
```

```python
print ('train R2: {} -- test R2: {}'.format(linreg.score(X_train_concat, y_train),
linreg.score(X_test_concat, y_test)))
```

```python
#The value 𝑅² = 1 corresponds to SSR = 0. That’s the perfect fit, since the values of predicted and actual responses fit completely to each other.
#it ranges from 0 to 1. Zero indicating that the proposed model does not improve prediction over the mean model, and One indicating perfect prediction. Better the model, higher the R-squared.

#In this particular case, I can say that the model is pretty nice because the R2_train is 0.618, so is pretty high.
```

```python
from sklearn.metrics import mean_squared_error

train_mse=mean_squared_error(linreg.predict(X_train_concat), y_train)
test_mse=mean_squared_error(linreg.predict(X_test_concat), y_test)
print ('train MSE: {} -- test MSE: {}'.format(train_mse, test_mse))
```

```python
from sklearn.metrics import mean_absolute_error as mae

train_mae=mae(linreg.predict(X_train_concat), y_train)
test_mae=mae(linreg.predict(X_test_concat), y_test)
print ('train MAE: {} -- test MAE: {}'.format(train_mae, test_mae))
```

```python
#It ranges from 0 to infinite, and lower value means better model. So, this model is a good model because the MAE value is almost 0.
```

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=100, k_neighbors=3)
X_train_SMOTE,y_train_SMOTE = sm.fit_resample(X_train_concat,y_train)
```

```python
X_train_SMOTE.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

LR = LogisticRegression(random_state=0, solver='lbfgs')
LR.fit(X_train_SMOTE, y_train_SMOTE)
pred = LR.predict(X_test_concat)

print("precision: ",precision_score(y_test,pred))
print("recall: ",recall_score(y_test,pred))
print("f1: ",f1_score(y_test,pred))
```

```python
#The most important rate is Recall, and in this case, the value is 0.9877 and i'ts almost 1, that it's mean that it's a good model.
#The precision score...The best value is 1 and the worst value is 0. In this case the value is pretty good.
#The F1 score can be interpreted as a harmonic mean of the precision and recall, where an F1 score reaches its best value at 1 and worst score at 0. The relative contribution of precision and recall to the F1 score are equal.
```
